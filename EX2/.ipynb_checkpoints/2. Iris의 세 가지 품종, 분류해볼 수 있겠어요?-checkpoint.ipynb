# Data split

sklearn에서는 다음과 같은 방법으로 간단하게 데이터 세트에서 train, test를 분류 할 수 있다. 

```python
from sklearn.model_selection import train_test_split
iris_label = iris.target
X_train,X_test,y_train,y_test = train_test_split(iris_data,
                                                iris_label,
                                                test_size=0.2,  # Data set에서 test set이 차지할 비율
                                                random_state=7) # =random seed
print("X_train 개수:",len(X_train),", X_test 개수:",len(X_test))
```

# 모델 사용하기

 sklearn에서는 간단한 모델들을 제공해준다. 여기서는 먼저 간단한 분류기인 Decision Tree를 사용해본다. Decision Tree는 sklearn.tree 패키지 안에 DecisionTreeClassifier라는 이름으로 내장되어 있다.
 학습과 예측등은 함수와 되어 있기 때문에 1~2줄로 끝난다.

## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

decision_tree = DecisionTreeClassifier(random_state=32)
print(decision_tree._estimator_type)
```

학습은 Classifier의 fit()이라는 method를 사용해 이루어진다.

```python
decision_tree.fit(X_train, y_train)
```

test데이터에서 예측을 할 때는 predict() method를 사용한다.

```python
y_pred = decision_tree.predict(X_test)
y_pred
```

정확도 측정을 위해서는 sklearn.metrics의 accuracy_score를 사용한다.

```python
from sklearn.metrics import accuracy_score

accuracy=accuracy_score(y_test,y_pred)
accuracy
```

## Random Forest

#### 랜덤 포레스트는 사람으로 표현하자면 회의를 하는 것이다. 회의와의 차이점은 서로 다른 요소들을 들고 의견을 낸다.
#### 서로 다른 요소를 쓰는 이유는 동일한 요소 set을 사용하게 되면 우선순위가 높은 속성이 거의 정해져 있기 때문이다. 각각은 서로 다른 요소를 가지고 Dicision 트리를 수행한다.
#### 이와같이 동일한 모델을 가지고 결과를 평균내는 방식들을 앙상블(Ensemble) 기법이라고 한다. Random forest는 이러한 이유로 sklearn.ensemble 패키지내에 존재한다.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

X_train, X_test, y_train, y_test = train_test_split(iris_data, 
                                                    iris_label, 
                                                    test_size=0.2, 
                                                    random_state=25)

random_forest = RandomForestClassifier(random_state=32)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)

print(classification_report(y_test, y_pred))
```

## Support Vector Machine (SVM)
##### SVM은 현재 데이터들을 다른 평면상에서 관점을 달리하여 보는 모델이다. xor같은 문제들은 선형모델로는 구분할 수 없는데 SVM과 같이 다른 평면상에서 보게 되면 비선형적인 관계를 만들 수 있게 된다.

```python
from sklearn import svm
svm_model = svm.SVC()

print(svm_model._estimator_type)



svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)

print(classification_report(y_test,y_pred))
```

## Logistic regression
#### 로지스틱 회귀 분석은 이진 분류를 수행하여 데이터 샘플을 양성(1) 또는 음성(0) 클래스 둘 중 어디에 속하는지 예측한다.

```python
from sklearn.linear_model import LogisticRegression
logistic_model = LogisticRegression()

print(logistic_model._estimator_type)


logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)

print(classification_report(y_test,y_pred))
```

## SGD(Stochastic Gradient Descent)
#### SGD는 확률적 경사하강법 알고리즘을 적용한 perceptron,logistic regression, SVM을 사용할 수 있게 해준다.
#### 아래와 같이 인자를 주지 않으면 defualt 값으로 설정된다. 많이 쓰이는 파라미터는 다음과 같다.

- #### `loss` : [preceptron, log(logistric regression), hinge(SVM)], Default = "hinge"

- #### `penalty(Regularization)` : [l2,l1,elasticnet], Default = "l2"

- #### `alpha(hyper param of penalty` : Default = "0.0001"

- #### `max_iter(epoch)` : Default = "1000" 

```python
from sklearn.linear_model import SGDClassifier
sgd_model = SGDClassifier()

print(sgd_model._estimator_type)

sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)

print(classification_report(y_test,y_pred))
```

# 정확도의 함정
#### 머신러닝에서는 모델을 학습하는 과정뿐 아니라 그 성능을 정확히 측정하는 과정이 중요하다. 어쩌면 모델을 학습하는 과정보다 더 중요할 것이다.
#### 이 정확도의 수치가 높다고 그대로 해석하는 것은 상당히 위험한데 이에 대한 사례를 MNIST 데이터를 사용해 확인해보자

```python
from sklearn.datasets import load_digits

digits = load_digits()
print(digits.keys())

digits_data = digits.data
print(digits_data.shape)

print(digits_data[0])
```

#### 먼저 MNIST의 데이터는 위와 같이 1797개의 이미지가 있는데 각 이미지들이 1차원으로 표현되어 있다. 이는 2차원 array를 1차원 array로 변경한 것이기 때문에 원래의 그림을 보기 위해서는 2차원으로 다시 reshape을 해야 한다.

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.imshow(digits.data[0].reshape(8, 8), cmap='gray')
plt.axis('off')
plt.show()
```

```python
print(digits.target)
print(digits.target_names)
```

#### 위와 같이 target의 이름을 보면 이 데이터는 0~9까지의 숫자들을 분류한 데이터라는 것을 확인할 수 있다.

#### 여기서 정확도의 함정을 확인하기 위해 변형을 가해 0~9까지를 분류하는게 아닌 3 인지 아닌지에 대한 이진 분류기를 만들 것이다.

```python
digits_label = digits.target
new_label = [3 if i == 3 else 0 for i in digits_label]
new_label[:20]
```

#### 이제 iris에서 한 것과 같이 Dcision tree를 이용해 분류를 하고 정확도를 측정해본다

```python

X_train,X_test,y_train,y_test = train_test_split(digits_data,
                                                new_label,
                                                test_size=0.2,  # Data set에서 test set이 차지할 비율
                                                random_state=15)

decision_tree = DecisionTreeClassifier(random_state=15)

decision_tree.fit(X_train, y_train)
y_pred = decision_tree.predict(X_test)

print(classification_report(y_test, y_pred))
```

```python
accuracy_score(y_test,y_pred)
```

#### 수치가 엄청 높게 나왔는데 사실 3에 대한 데이터와 나머지에 대한 데이터의 비율이 1:9이므로 모든 답이 0일 경우 정확도는 90%가 보장되어 있다. 다음과 같이 모든 label이 0으로 되어있는 데이터로 학습을 하면 90%이상의 정확도가 나온다.

```python
fake_pred = [0] * len(y_pred)

accuracy = accuracy_score(y_test, fake_pred)
accuracy
```

# Confusion matrix

![](../img/confusion_matrix.png)

#### 위에서 확인 정확도는 전체 데이터 중 맞은 데이터에만 신경쓰는 척도이다. 하지만 실제로는  task에 따라 오답을 최소화 하는게 목적일 수도 있고 정답을 최대화 하는게 목적일 수도 있다. 이러한 정답과 오답을 구분하여 표현하는 방법을 오차행렬이라고 한다.
#### 오차행렬의 요소는 4가지로 구분이 되는데 Positive와 Negative가 모델에서 예측한 대답(1과 0), True와 False는 모델에서 예측한 이 대답이 정답인지 아닌지를 판단해 구분한다.
#### 방금 MNIST에서 예측한 두 결과의 confusion matrix를 확인해보자

```python
from sklearn.metrics import confusion_matrix

print(confusion_matrix(y_test,y_pred))
print(confusion_matrix(y_test,fake_pred))
```

#### 위의 결과를 보면 fake_pred에서는 Negative로 예측한 것은 없기 때문에 FN과 TN 모두 0이 된다.

#### 위에서 계속 써왔던 classification_report에는 이에 대한 정보가 나와있다.

```python
print(classification_report(y_test, y_pred))
print(classification_report(y_test, fake_pred))
```

#### 맨 처음 결과는 0.58의 precision과 0.67의 recall을 갖고 2번째 결과는 모두 0의 precision과 recall을 갖는다. 반면 정확도는 0.94와 0.93으로 비슷하다. 이는 데이터의 불균형으로 인한 결과이다.
#### 참고로 recall은 sensitivity값에 해당한다.

# MNIST

```python
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd

digits = load_digits() # data download
digits_data = digits.data # data of digits(image's pixel value)
digits_label = digits.target # label of digits
pd_digits = pd.DataFrame(data=digits_data,columns =digits.feature_names)
pd_digits['label']=digits_label
pd_digits
```

```python
# Data split
X_train,X_test,y_train,y_test = train_test_split(digits_data,
                                                digits_label,
                                                test_size=0.2,  
                                                random_state=15)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn import svm
from sklearn.linear_model import SGDClassifier
from sklearn.linear_model import LogisticRegression

# Decision Tree
decision_tree = DecisionTreeClassifier(random_state=15)
decision_tree.fit(X_train,y_train)
y_pred = decision_tree.predict(X_test)
print("Decision Tree\n",classification_report(y_test, y_pred))

# Random Forest
random_forest = RandomForestClassifier(random_state=15)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
print("Random Forest\n",classification_report(y_test, y_pred))

# Support Vector Machine
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print("Support Vector Machine\n",classification_report(y_test,y_pred))

# Stochastic Gradient Descent
sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
print("Stochastic Gradient Descent\n",classification_report(y_test,y_pred))

# Logistic Regression
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
print("Logistic Regression\n",classification_report(y_test,y_pred))
```

현재 예제에서는 svm이 precision, recall 평가 값이 모두 높아서 svm모델을 하는게 가장 좋다. 
만약 precision과 recall이 극과 극으로 나왔다면 , 개인적으로는 precision이 높아야 한다고 생각한다. 아래 사진은 precision과 recall의 값에 따른 이미지의 분류인데 precision이 높을 때 잘못 분류된 사항을 보면 사람이어도 착각할 수도 있겠다 싶은 사진이 나타난다. 반면 재현율에 가중치를 뒀을 때는 의아한 사진이 점점 등장함을 볼 수 있다.

# Wine

```python
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report
import pandas as pd

wine = load_wine() # data download
wine_data = wine.data # data of digits(image's pixel value)
wine_label = wine.target # label of digits
pd_wine = pd.DataFrame(data=wine_data,columns =wine.feature_names)
pd_wine['label']=wine_label
pd_wine
```

```python
# Data split
X_train,X_test,y_train,y_test = train_test_split(wine_data,
                                                wine_label,
                                                test_size=0.2,  
                                                random_state=15)
```

```python
# Decision Tree
decision_tree = DecisionTreeClassifier(random_state=15)
decision_tree.fit(X_train,y_train)
y_pred = decision_tree.predict(X_test)
print("Decision Tree\n",classification_report(y_test, y_pred))

# Random Forest
random_forest = RandomForestClassifier(random_state=15)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
print("Random Forest\n",classification_report(y_test, y_pred))

# Support Vector Machine
svm_model = svm.SVC()
svm_model.fit(X_train, y_train)
y_pred = svm_model.predict(X_test)
print("Support Vector Machine\n",classification_report(y_test,y_pred))

# Stochastic Gradient Descent
sgd_model = SGDClassifier()
sgd_model.fit(X_train, y_train)
y_pred = sgd_model.predict(X_test)
print("Stochastic Gradient Descent\n",classification_report(y_test,y_pred))

# Logistic Regression
logistic_model = LogisticRegression()
logistic_model.fit(X_train, y_train)
y_pred = logistic_model.predict(X_test)
print("Logistic Regression\n",classification_report(y_test,y_pred))
```

위의 결과를 보면 이상하게 SVM과 SGD(default 이므로 SVM사용)의 결과가 현저하게 낮은데 feature의 범위가 상이해서 나온 결과이다.
아래의 값을 보면 alcohol은 10^1단위 이고 magnesium은 10^1\~10^2,proline은 10^2\~10^3 범위를 갖는다.
SVM은 이러한 input의 scale에 민감하므로 전처리를 해야한다.

```python
pd_wine
```

```python
from sklearn.preprocessing import MinMaxScaler


mms_scaler = MinMaxScaler()
mms_scaler.fit(X_train)

x_train_scaled = mms_scaler.transform(X_train)
x_test_scaled = mms_scaler.transform(X_test)

# Support Vector Machine
svm_model = svm.SVC()
svm_model.fit(x_train_scaled, y_train)
y_pred = svm_model.predict(x_test_scaled)
print("Support Vector Machine\n",classification_report(y_test,y_pred))

# Stochastic Gradient Descent
sgd_model = SGDClassifier()
sgd_model.fit(x_train_scaled, y_train)
y_pred = sgd_model.predict(x_test_scaled)
print("Stochastic Gradient Descent\n",classification_report(y_test,y_pred))
```

#### 전처리 하나만으로 정확도가 올라간 것을 확인할 수 있다.

#### 결과를 보면 Random Forest와 SGD의 결과가 precision과 recall 둘다 100으로 높다.
#### 만약 실제로 와인을 구매한다고 했을 때는 높은 등급의 와인을 샀는데 낮은 등급의 와인이 오면 손해이므로 precision이 높아야 좋고 판매자 입장에서는 높은 등급이 낮은 등급으로 팔리면 안되므로 recall이 중요하지 않을까 생각한다. 물론 둘다 높으면 좋다고 생각한다.

# Breast_cancer

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

cancer = load_breast_cancer() # data download
cancer_data = cancer.data # data of digits(image's pixel value)
cancer_label = cancer.target # label of digits
pd_cancer = pd.DataFrame(data=cancer_data,columns =cancer.feature_names)
pd_cancer['label']=cancer_label
pd_cancer
```

```python
# Data split
X_train,X_test,y_train,y_test = train_test_split(cancer_data,
                                                cancer_label,
                                                test_size=0.2,  
                                                random_state=15)
```

#### 이번에도 feature들의 scale이 모두 상이하므로 전처리를 해서 진행을 할 것이다.

```python
from sklearn.preprocessing import MinMaxScaler


mms_scaler = MinMaxScaler()
mms_scaler.fit(X_train)

x_train_scaled = mms_scaler.transform(X_train)
x_test_scaled = mms_scaler.transform(X_test)

# Decision Tree
decision_tree = DecisionTreeClassifier(random_state=15)
decision_tree.fit(x_train_scaled,y_train)
y_pred = decision_tree.predict(x_test_scaled)
print("Decision Tree\n",classification_report(y_test, y_pred))

# Random Forest
random_forest = RandomForestClassifier(random_state=15)
random_forest.fit(x_train_scaled, y_train)
y_pred = random_forest.predict(x_test_scaled)
print("Random Forest\n",classification_report(y_test, y_pred))

# Support Vector Machine
svm_model = svm.SVC()
svm_model.fit(x_train_scaled, y_train)
y_pred = svm_model.predict(x_test_scaled)
print("Support Vector Machine\n",classification_report(y_test,y_pred))

# Stochastic Gradient Descent
sgd_model = SGDClassifier()
sgd_model.fit(x_train_scaled, y_train)
y_pred = sgd_model.predict(x_test_scaled)
print("Stochastic Gradient Descent\n",classification_report(y_test,y_pred))

# Logistic Regression
logistic_model = LogisticRegression()
logistic_model.fit(x_train_scaled, y_train)
y_pred = logistic_model.predict(x_test_scaled)
print("Logistic Regression\n",classification_report(y_test,y_pred))
```

#### 이 데이터의 경우 실제 암인데 암이 아닌걸로 분류되면 안되므로 1에 대한 recall이 가장 중요하다. 따라서 Logistic Regression의 성능이 100으로 가장 좋다고 판단할 수 있다.
